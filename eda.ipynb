```python
bins = [0,100000,300000 , float('inf')]
labels = ['Low', 'Medium', 'High']
df['Income_band'] = pd.cut(df['Sum of Estimated Income'], bins=bins, labels=labels , right = False)
```

```python
#examine the distribution of unquie categories in ctaegorical column
categorical_cols = df[["Sum of BRId","Sum of GenderId","Sum of IAId","Sum of Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification", "Sum of Properties Owned" , "Sum of Risk Weighting","Income_band"]].columns
for col in categorical_cols:
  print(f"Value Counts for {col}: ")
  display(df[col].value_counts())
  print("\n")
```

```python
for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(9,4))
  sns.histplot(df[col])
  plt.title(f'Histogram of {col}')
  plt.xlabel(col)
  plt.ylabel('count')
  plt.show()
```

```python
numerical_cols =['Sum of Estimated Income','Sum of Superannuation Savings']

plt.figure(figsize=(14,10))
for i , col in enumerate(numerical_cols):
  plt.subplot(2,1,i+1)
  sns.histplot(df[col])
  plt.title({col})
plt.show()
```

```python
numerical_cols =["Sum of Estimated Income","Sum of Superannuation Savings","Sum of Credit Card Balance","Sum of Bank Loans","Sum of Bank Deposits","Sum of Checking Accounts","Sum of Saving Accounts","Sum of Foreign Currency Account","Sum of Business Lending"]

plt.figure(figsize=(14,10))
for i ,col in enumerate(numerical_cols):
  plt.subplot(3,3,i+1)
  sns.histplot(df[col],kde =True)
  plt.title(f'Histogram of {col}')
  plt.xlabel(col)
plt.tight_layout()
plt.show()
```

```python
numerical_cols =["Sum of Estimated Income","Sum of Superannuation Savings","Sum of Credit Card Balance","Sum of Bank Loans","Sum of Bank Deposits","Sum of Checking Accounts","Sum of Saving Accounts","Sum of Foreign Currency Account","Sum of Business Lending"]
correlaion_matrix =df[numerical_cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlaion_matrix,annot=True,cmap='coolwarm',fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
